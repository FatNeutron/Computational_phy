## Bisection method


```python
import math as m

def f(x):
    return m.sin(x)

def bisection(x, y, f):
    a, b = x, y
    tol = 1.0 * m.pow(10, -6)
    itera = 20

    if abs(f(a)) < tol:
        print("The root is %.8f (at a)" % a)
        return
    elif abs(f(b)) < tol:
        print("The root is %.8f (at b)" % b)
        return
    elif f(a) * f(b) > 0:
        print("This is a bad interval.")
        return
    else:
        runs = 1
        while runs < itera:
            mid = (a + b) / 2  # Calculate mid here
           

            print("%2d \t %.8f \t %.8f \t %.8f \t %.8f" % (runs, a, b, mid, f(mid)))

            if abs(f(mid)) < tol:  # Check for convergence
                y0 = m.pi
                print("The root is %.8f" % mid)
                print("deviation is %.8f"%(mid - y0))
                return

            if f(a) * f(mid) < 0:
                b = mid
            else:
                a = mid

            runs += 1

        print("Could not converge in %d iterations." % itera)

# Example usage
bisection(3, 3.2, f)
```
