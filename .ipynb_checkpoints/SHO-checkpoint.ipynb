### Calulating $(a + a^\dagger)^3$

```python
from sympy import symbols, expand, Function, pprint, init_printing, latex
from sympy import conjugate, I
from IPython.display import display, Math  # Import display from IPython

# Enable pretty printing (LaTeX style)
init_printing()

# Define annihilation (a) and creation (a†) operators as noncommutative symbols
a = symbols('a', commutative=False)
adag = symbols(r'a^\dagger', commutative=False)
# Define the expression
expr = (a + adag)**3

alpha = symbols(r'\alpha')

# Expand the expression
expr = expand(expr)

# Incorporate the commutator relation [a, a†] = 1
commutator = {a*adag: adag*a + 1}  # Rewriting a*adag in terms of adag*a using the commutator

sub1 = {a: alpha, adag: conjugate(alpha)}
# Replace using the commutator relation
expr = expand(expr).subs(commutator)
expr = expand(expr).subs(commutator)
expr = expand(expr).subs(commutator)
expr = expand(expr).subs(sub1)
# Display the result in LaTeX format
display(Math(latex(expr)))  # Ensure LaTeX output looks correct
```
