```python
import numpy as  np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

## Definitions

```python
def euler(f, x, y, h):
    
    return y + h * f(x, y)

def midpoint(f, x, y, h):
    
    k1 = h * f(x, y)
    return y + h * f(x + h / 2, y + k1 / 2)

def rk4(f, x, y, h):
    
    k1 = h * f(x, y)
    k2 = h * f(x + h / 2, y + k1 / 2)
    k3 = h * f(x + h / 2, y + k2 / 2)
    k4 = h * f(x + h, y + k3)
    return y + (k1 + 2 * k2 + 2 * k3 + k4) / 6


def caller(my_method, fn, y_ini, N, xs, h):
    y = y_ini
    ys = np.zeros((N, len(y_ini)), dtype=np.float64)
    for i in range(N):
        x = xs[i]
        ys[i, :] = y
        y = my_method(fn, x, y, h)
    return ys
```

### Question 1

```python
# differential eq : dx/dt = -xt        # here y = x and x = t

def f(x,t):
    return -x*t 

# parameters

h = 0.01

t_init = 0
t_final = 15
x_0 = 1


N  = int(abs(t_final - t_init)/h)

ts = np.arange(t_init,t_final,h)
x_ini = np.array([x_0])

xE = caller(euler,f,x_ini,N,ts,h)
xM = caller(midpoint,f,x_ini,N,ts,h)


```

```python
tfine = np.linspace(t_init,t_final,1000)
xfine = np.exp(-tfine**2 /2)
xexact = np.exp(-ts**2 /2)


plt.figure(figsize=(6, 6))
plt.plot(tfine, xfine, color='blue', linewidth=1, label='Exact')


plt.scatter(ts, xE, label='Euler')
plt.scatter(ts,xM,label = "midpoint")
plt.legend()
plt.title("Comparison of Numerical Methods with Exact Solution")
plt.grid()
plt.tight_layout()
```

```python
hs = 10.0 ** np.arange(-4.0, -2.0 + 0.2, 0.2)

t_init = 0
t_final = 1.5
x0 = 0.0
y_ini = np.asarray([1.0])

errors1 = np.zeros(len(hs))
errors2 = np.zeros(len(hs))
for i, h in enumerate(hs):
    xs = np.arange(x0, 1.5 + h, h) # Generate xs up to 1.5 with step h
    N = len(xs)
    ys1 = caller(euler, f, y_ini, N, xs, h)
    ys2 = caller(midpoint, f, y_ini, N, xs, h)

    errors1[i] = abs(ys1[-1] - np.exp(-(xs[-1])**2 /2))
    errors2[i] = abs(ys2[-1] - np.exp(-(xs[-1])**2 /2))

    
print("for euler errors:",errors1)
print("\n")
print("for midpoint errors:",errors2)

```

```python
# Error data in log-log format
xd = np.log10(hs)
yd1 = np.log10(errors1)
yd2 = np.log10(errors2)


# Define the fitting function
def l(x, p1, p2):
    return p1 * x + p2


# Fit the data
params1,cov1 = curve_fit(l, xd, yd1)

# Fit the data
params2,cov2 = curve_fit(l, xd, yd2 )

# Generate smooth fit
xfine = np.linspace(xd[0], xd[-1], 1000)
yfine1 = l(xfine, *params1)
yfine2 = l(xfine,*params2)


# Plot the data and the fit
plt.figure(figsize=(8, 6))
plt.scatter(xd, yd1, label="Data")
plt.plot(xfine, yfine1, label=f"Fit. Slope = {params1[0]:.2f}", color="orange")
plt.xlabel(r"$\log_{10}$ of h")
plt.ylabel(r"$\log_{10}$ of Errors")
plt.legend(loc="upper left")
plt.grid()
plt.title("Log-Log Error(euler) vs. Step Size")






# Plot the data and the fit
plt.figure(figsize=(8, 6))
plt.scatter(xd, yd2, label="Data")
plt.plot(xfine, yfine2, label=f"Fit. Slope = {params2[0]:.2f}",c='r')
plt.xlabel(r"$\log_{10}$ of h")
plt.ylabel(r"$\log_{10}$ of Errors")
plt.legend(loc="upper left")
plt.grid()
plt.title("Log-Log Error Midpoint vs. Step Size")

plt.tight_layout()
plt.show()



print("since slope is more the midpoint is best")
```

```python
# function dy/dt = p
# function dp/dt = -4pi2y

def g(t,y):
    return np.asarray([y[1], -4*(np.pi)**2 * y[0]])



h = 0.01

t_init = 0
t_final = 15



N  = int(abs(t_final - t_init)/h)

ts = np.arange(t_init,t_final,h)
x_ini = np.asarray([1.0,0.0])

xE = caller(euler,g,x_ini,N,ts,h)
xM = caller(midpoint,g,x_ini,N,ts,h)

xexact = np.linspace(0,15,1000)

# Plot for Euler method
plt.figure(figsize=(8, 10))
plt.subplot(3, 1, 1)
plt.plot(ts, xE[:, 0], label="Position")
plt.plot(ts, xE[:, 1], label="Velocity")


plt.plot(xexact,np.cos(2*np.pi*xexact))

plt.plot(xexact,-np.sin(2*np.pi*xexact))
plt.title("Euler")
plt.legend()
plt.xlabel("Time")
plt.ylabel("Value")
plt.grid()


# Plot for Midpoint method
plt.subplot(3, 1, 2)


plt.plot(xexact,np.cos(2*np.pi*xexact))
plt.plot(xexact,-np.sin(2*np.pi*xexact))
plt.plot(ts, xM[:, 0], label="Position")
plt.plot(ts, xM[:, 1], label="Velocity")
plt.title("Midpoint")
plt.legend()
plt.xlabel("Time")
plt.ylabel("Value")
plt.grid()

plt.tight_layout()


#plot for exact 



```
