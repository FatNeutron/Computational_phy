# Importance Sampling

```python
import  numpy as np
import matplotlib.pyplot as plt
```

## Question 1

```python
def simpsons(ys,h):
    return (h / 3) * (ys[0] + ys[-1] + 4 * np.sum(ys[1:-1:2]) + 2 * np.sum(ys[2:-1:2]))      # i used simpsons for reference areas
```

```python
import numpy as np

def f(x):
    return np.where(x == 0, 0.0, x**3 / (np.exp(x) - 1))     # Use np.where to apply the limit value at x = 0 (L Hopital rule)
                                                            # np.where(condition , value if true , value if false)
def w(x):
    norm = (1/np.sqrt(b)) * (np.arctan((1 - x0)/np.sqrt(b)) - np.arctan((0-x0)/np.sqrt(b)))
    return  (1/norm)*(1/( b + (x-x0)**2))

b = 0.5
x0 = 1.2

xs,h = np.linspace(0,1,1000,retstep = True)
ys_f = f(xs) 
ys_w = w(xs)
print(f'The area under the w(x) curve is :{simpsons(ys_w,h)}\nso w(x) is valid PDF')

################### Plotting #########################
fig, ax = plt.subplots(1, 2, figsize=(8, 4))

ax[0].plot(xs,ys_f)
ax[0].set_title("$f\,(x)$")
ax[0].set_xlabel('x')
ax[0].set_ylabel('y_f')
ax[0].grid()

ax[1].plot(xs,ys_w,c='orange')
ax[1].set_title("$w\,(x)$")
ax[1].set_xlabel('x')
ax[1].set_ylabel('y_w')
ax[1].grid()
```

```python
# lets generate the CDF (area upto x) :
#  y(x) = (1/root_b) * (np.arctan((x - x0)/root_b) - np.arctan((-x0)/root_b))
# lets inverse y(x) 
# so when we do random in y it gives uniform area values but corresponding x will follow w(x) distribution. since uniform areas
# might be genrated because of x that are very close near the peak so more points from the peak region 

def getx(y):
    return ((np.sqrt(b))*(np.tan( (y*np.sqrt(b)) + np.arctan(0-x0/np.sqrt(b))  ))) + x0

def g(x):
    return  f(x)/w(x)
```

```python
def MC_eval_f(n):
    store = 0.0
    for i in range(n):
        x = np.random.rand()            # normal sampling using uniform random generator 
        store += f(x)
    return store/n

def MC_eval_g(n):
    store = 0.0
    for i in range(n):
        y = np.random.rand()           # important sampling using biased distribution random generator
        store += g(getx(y))
    return store/n
```

```python
plt.figure(figsize = (10,6))

xs,h = np.linspace(0,1,1000,retstep = True)
ys = f(xs)
original_integral = simpsons(f(xs),h)            # finding original integral using simpsons rule
plt.plot(xs, ys, label='f(x)')

ys = w(xs)
plt.plot(xs, ys, label='w(x)')
ys = g(xs)
plt.plot(xs, ys, label='f(x)/w(x)')

plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.legend();
```

```python
Ns = np.array([10,20,30,50,100,200])
np.random.seed(1)
print("    N\tMC_rand\t     Err\tMC_Impo\t   Err")
for i in Ns:
    normal  = MC_eval_f(i)
    impor   = MC_eval_g(i)
    print('%5d %10.7f %10.7f %10.7f,%10.7f '%(i,normal,abs(original_integral-normal),impor,abs(original_integral-impor)))
print("\n\nthe original area is",original_integral)   
print(f'\nClearly we see that important sampling gives better results at even less data points')
```

## Question 2

```python
def w(x):
    return 1/(1+(x-1)**2)

xs = np.linspace(-15,20,10000)
ys = w(xs)    # already normalized as max = 1
plt.figure(figsize=(8,4)) 
plt.axvline(x=1, linestyle='--',c = "red")   # peak at x = 1
plt.plot(xs,ys)
plt.title("w(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
```

```python
x0 = 0.5        # starting point
N = int(1e+6)    # Max No of random walks
stepsize = 1.0  
np.random.seed(12)
positions = np.zeros(N)
for i in range(N):
    positions[i] = x0        # so the first position is always x0 that we gave
    x1 = x0 + stepsize*(2*np.random.rand()-1)         # it can be between -1 to 1 
    r = w(x1)/w(x0)
    if r>1:
        x0 = x1                          # updating or goint to next step as function is increasing 
    else:
        p = np.random.rand()
        if p <= r:
            x0 = x1           # this condition is very less often so it mostly stays at previous step as it is closer to peak.
 

 # 1000 bins  or 1000 groups  and values = no of times the position repeated in each bin 
#  edges are the starting  and ending  points of each bins 
stay_counts, edges = np.histogram(positions,1000)

# if N bins are there there will be N+1 edges as the last one also counts
midpoints = [(edges[i]+edges[i+1])/2 for i in range(len(edges)-1)]          # midpoint of those bins for efficient plotting 

# normalization 
stay_counts = stay_counts/np.max(stay_counts)  # stay counts means how many times did they stay in that position for a points in bins

plt.figure(figsize=(8,4))
plt.axvline(x=1 , linestyle = "--",c = "r")               # expected peak at x = 1
plt.plot(midpoints, stay_counts,c='orange',label='Generated')
plt.xlim([-16,21])
plt.grid()
plt.title("Generated Normalized distribution")
plt.legend();

# stay_counts are proportional to W(x) so normalized graph will follow w(x) distribution
```
