##  Velocity Verlet

```python
import numpy as np
import matplotlib.pyplot as plt


def ax(t,x,y):
    return -x/(x**2 + y**2)**(3/2)

def ay(t,x,y):
    return -y/(x**2 + y**2)**(3/2)


def vel_verlet(f,g,t_i,t_f,cond,h):
    
    ts = np.arange(t_i,t_f,h)
    xs = np.zeros(len(ts))
    ys = np.zeros(len(ts))
    vxs = np.zeros(len(ts))
    vys = np.zeros(len(ts))
    energy = np.zeros(len(ts))
    
    xs[0],ys[0],vxs[0],vys[0],energy[0] = cond
    

    for i,tn in enumerate(ts[:-1]):
        xs[i+1] = xs[i] + vxs[i]*h + 0.5*f(tn,xs[i],ys[i])*h**2
        ys[i+1] = ys[i] + vys[i]*h + 0.5*g(tn,xs[i],ys[i])*h**2
        vxs[i+1] = vxs[i] + 0.5*h*(f(tn,xs[i],ys[i]) + f(tn+h,xs[i+1],ys[i+1]))
        vys[i+1] = vys[i] + 0.5*h*(g(tn,xs[i],ys[i]) + g(tn+h,xs[i+1],ys[i+1]))
        
        energy[i+1] = 0.5*(vxs[i+1]**2 + vys[i+1]**2) - 1/(xs[i+1]**2 + ys[i+1]**2)**0.5
        
    return ts,xs,ys,vxs,vys,energy



cond = [1,0,0,1, (0.5 - 1)]
ts,xs,ys,vxs,vys,energy = vel_verlet(ax,ay,0,10,cond,0.001)

fig,ax = plt.subplots(figsize = (6,6))

ax.plot(xs,ys)


```

```python
pos = np.where(np.abs(ys)<0.001)
close_zero = ys[pos]
time = ts[pos]
time
```

```python
theta = np.zeros(len(ts))

for i  in range(len(theta)):
    theta[i] = np.arctan(ys[i]/xs[i])

pos = np.where(np.abs(theta) < 0.001)
time = ts[pos]
time
```
